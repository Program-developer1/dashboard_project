```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set_theme(style='whitegrid')
%matplotlib inline
```

```python
df = pd.read_csv('../data/2_5_month.csv')
df['time'] = pd.to_datetime(df['time'], utc=True)
df['date'] = df['time'].dt.date
print('Shape:', df.shape)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
print('Missing values:')
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].hist(df['mag'].dropna(), bins=30, color='#1a6faf', edgecolor='white')
axes[0].set_title('Magnitude Distribution')
axes[0].set_xlabel('Magnitude')
axes[1].hist(df['depth'].dropna(), bins=30, color='#2196F3', edgecolor='white')
axes[1].set_title('Depth Distribution')
axes[1].set_xlabel('Depth (km)')
plt.tight_layout()
plt.show()
```

```python
daily = df.groupby('date').size().reset_index(name='count')
daily['date'] = pd.to_datetime(daily['date'])
plt.figure(figsize=(12, 4))
plt.plot(daily['date'], daily['count'], color='#1a6faf', marker='o', markersize=3)
plt.title('Daily Earthquake Count')
plt.xlabel('Date')
plt.ylabel('Count')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
cols = ['mag', 'depth', 'latitude', 'longitude', 'gap', 'rms']
corr = df[cols].dropna().corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='Blues')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
print('Top 10 locations:')
print(df['place'].str.extract(r',\s*(.+)$')[0].value_counts().head(10))
```

## Key Insights
- **1,775 events** recorded over ~2.5 months worldwide
- Average magnitude: **3.7**, Max magnitude: **6.9**
- Most earthquakes are shallow (< 70 km depth)
- Alaska, Western US, and Pacific regions are most active
- `mb` and `ml` are the most common magnitude types
- `us` (USGS) network monitors the majority of global events
